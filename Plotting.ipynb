## <p style="text-align: center;"> <font size = 6>PLOTTING GRAPHS USING MATPLOTLIB</font><p> <hr></hr>
  To import matplotlib we can write
  - <span style="color:blue">import</span> matplotlib.pyplot <span style="color:blue">as</span> plt
  - <span style="color:blue">from</span> matplotlib.pyplot <span style="color:blue">import</span> pyplot, show

  
  When you import pyplot as plt, you have access to all functions and objects within the pyplot module using the plt. prefix. 
    
  On the other hand, when you import specific functions like **plot()** and **show()** directly, you can use them without the need for the plt. prefix, but you won't have access to other functions and objects from the pyplot module unless you import them separately.

  We have 2 ways of plotting with matplotlib:
  - Pyplot 
  - Object-oriented

## Pyplot Approach
The pyplot approach is a more concise and convenient way of creating plots.

Here you directly call the plotting functions from the pyplot module, and Matplotlib internally manages the Figure and Axes objects for you (no need of creating fig & ax objects)

```python
from matplotlib.pyplot import plot, xlabel, ylabel, title, figure, show

# Create some data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Set the figure size
fig = figure(figsize=(3, 3))
ax = fig.add_subplot(111)

# Plot the data
title("Plotting x against y")
xlabel("X-axis(units)")
ylabel("Y-axis(units)")
plot(x, y)
# Show the plot
show()
```

# <p style="text-align: center;"> <font size = 7 >OR</font><p>                 

```python
import matplotlib.pyplot as plt
# Create some data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Plot the data
plt.title("Plotting x against y")
plt.xlabel("X-axis(units)")
plt.ylabel("Y-axis(units)")
plt.plot(x, y)
# Show the plot
plt.show()
```

## Object-oriented way
   <p style = "text-align: center;"> Syntax: <b>fig, ax = plt.subplots(nrow, ncol, figsize) </b> <p>
    
- 1st object **fig**, short for figure, imagine it as the frame of your plot that you can reframe but you cannot draw on it. 
- 2nd object, **ax**, short for axes, is the canvas you draw on. Or rephrasing, it is the blank sheet you can plot and hold your data. An axes object can only belong to one figure.
  Remember axis i.e. **ax**  is an object within the figure i.e. **fig** object
  


## Example

Let's try creating more in 1 figure, we then have 2 methods of accessing each individual plot i.e. 

a) <span style="color:red">**Indexing**</span>

```python
fig, ax = plt.subplots(1,2,figsize=(10,5))  # - Short Form

#fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(10,5))  - Long form
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [10, 8,6,4,2]

ax[0].plot(x,y) 
ax[1].plot(x,z)
#Adjusting the spacing using tight_layout, controls spacing inbtwn, gives breathing room
fig.tight_layout(pad=7)
```

 Here you are indexing the ax array to access individual subplots by their index positions. Here, you are explicitly specifying the index of the subplot you want to plot on.

b)<span style="color:red"> **Unpacking** </span>

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5))
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
z = [10, 8,6,4,2]
ax1.plot(x,y)
ax2.plot(x,z)
```

Here you are "unpacking" the individual Axes objects and directly plotting on them. Here, you are not using indexing but rather referencing the individual Axes objects directly.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create some sample data
x = np.random.randn(100)
y1 = np.random.randn(100)
y2 = np.random.randn(100)

# Create a figure and Axes objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

fig.tight_layout(pad=7)

# Plot scatter plot on the first Axes
ax1.scatter(x, y1, color='red', label='Data 1')
ax1.scatter(x, y2, color='blue', label='Data 2')

# Customize the first plot
ax1.set_title('Scatter Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.legend()

# Plot histogram on the second Axes
ax2.hist(x, bins=10, color='green')

# Customize the second plot
ax2.set_title('Histogram')
ax2.set_xlabel('Values')
ax2.set_ylabel('Frequency')

# Show the plots
plt.show()
```

## Note
Here is something important that you can add to your arsenal of plots -> <span style = 'color:blue'> You can use pandas library to use plots </span>

 See, pandas library includes a wrapper for matplotlib that will let you use a <span style = 'color:red'>.plot() </span> method on a data frame, e.g.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample DataFrame
data = {'x': [1, 2, 3, 4, 5], 'y': [10, 20, 15, 25, 30]}
df = pd.DataFrame(data)

# Plot a line plot
df.plot(x='x', y='y', kind='line')

# Plot a bar plot
df.plot(x='x', y='y', kind='bar')

# Plot a scatter plot
df.plot(x='x', y='y', kind='scatter')
plt.show()
```
